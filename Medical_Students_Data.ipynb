#Medical_Students_Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("https://github.com/Taj1920/Data-Analysis-Datasets/raw/refs/heads/main/Datasets-Cleaning/medical_students_dataset.csv",encoding="ISO-8859-1")
df.info()
```

```python
df['Student ID'].unique()
```

```python
df["Age"].unique()
```

```python
#Handling Age
df["Age"]=df["Age"].replace(np.nan,0)
df["Age"]=df["Age"].astype(float)
avg_age=round(df["Age"].mean())
df["Age"]=df["Age"].replace(0,avg_age)
df["Age"].unique()
```

```python
df["Gender"].unique()
df["Gender"] = df["Gender"].fillna("Not disclosed")
df["Gender"].unique()
```

```python
df["Height"].unique()
```

```python
df["Weight"].unique()
```

```python
df["Blood Type"].unique()
```

```python
#Handling Blood Type
df["Blood Type"] = df["Blood Type"].fillna(df["Blood Type"].mode()[0])
df["Blood Type"].unique()
```

```python
df["BMI"].unique()
```

```python
#Handling BMI
df["BMI"] = df["BMI"].fillna(df["BMI"].median())
df["BMI"].unique()
```

```python
df['Temperature'].unique()
```

```python
#Handling Temperature
df["Temperature"] = df["Temperature"].fillna(df["Temperature"].mean())
df["Temperature"].unique()
```

```python
df["Heart Rate"].unique()
```

```python
#Handling Heart Rate
df["Heart Rate"] = df["Heart Rate"].fillna(df["Heart Rate"].median())
df["Heart Rate"].unique()
```

```python
df["Blood Pressure"].unique()
```

```python
#Handling Blood Pressure
df["Blood Pressure"] = df["Blood Pressure"].fillna(df["Blood Pressure"].median())
df["Blood Pressure"].unique()
```

```python
df["Cholesterol"].unique()
```

```python
#Handling Cholesterol
df["Cholesterol"] = df["Cholesterol"].fillna(df["Cholesterol"].median())
df["Cholesterol"].unique()
```

```python
df["Diabetes"].unique()
```

```python
#Handling Diabetes
df["Diabetes"] = df["Diabetes"].fillna(df["Diabetes"].mode()[0])
df["Diabetes"].unique()
```

```python
df["Smoking"].unique()
```

```python
#Handling Smoking
df["Smoking"] = df["Smoking"].fillna(df["Smoking"].mode()[0])
df["Smoking"].unique()
```

```python
df.shape
```

```python
#Dropping Duplicates
df.drop_duplicates(inplace=True)
df.shape
```

```python
#6)Handling missing values
df.dropna(inplace=True)
df.isnull().sum()
```

```python

#10) Resetting the index
df.reset_index(inplace=True)
df=df.drop("index",axis=1)
df
```
