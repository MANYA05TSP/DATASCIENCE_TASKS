```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://github.com/Taj1920/Data_Analysis/raw/refs/heads/main/Youtube_analysis/YOUTUBE%20CHANNELS%20DATASET.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Ranking"].unique()
```

```python
df["Ranking"]=df["Ranking"].fillna(101)
df["Ranking"].unique()
```

```python
df["Username"].unique()
```

```python
df["Username"]=df["Username"].fillna("Unknown")
df["Username"].unique()
```

```python
df["Subscribers"].unique()
```

```python
df["Subscribers"]=df["Subscribers"].str.strip('M')
```

```python
df["Subscribers"]=df["Subscribers"].astype(float)
# df["Subscribers"]=df["Subscribers"].astype(int)
df["Subscribers"]=df["Subscribers"].fillna(round(df["Subscribers"].mean()))
df["Subscribers"].unique()
```

```python
df["Uploads"].unique()
```

```python
df["Uploads"]=df["Uploads"].fillna("0")
df["Uploads"].unique()
```

```python
df["Views"].unique()
```

```python
df["Views"]=df["Views"].fillna("0")
```

```python
df["Views"]=df["Views"].str.replace(',',"")
df["Uploads"]=df["Uploads"].str.replace(',',"")
df["Views"].unique()
```

```python
df["Country"].unique()
```

```python
df["Country"]=df["Country"].fillna("Unknown")
df["Country"].unique()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df
```

```python
#uploads vs subscribers
plt.scatter(df["Uploads"], df["Subscribers"])
plt.xlabel("Uploads")
plt.ylabel("Subscribers")
plt.title("Uploads vs Subscribers")
plt.show()
```

```python
#Channels by country
df["Country"].value_counts().plot(kind="bar")
plt.title("Number of Channels by Country")
plt.show()
```

```python
#Country share
df["Country"].value_counts().head(5).plot(kind="pie", autopct='%1.1f%%')
plt.title("Top Countries on YouTube")
plt.show()
```

```python
#Ranking vs Subsribers
plt.plot(df["Ranking"], df["Subscribers"])
plt.xlabel("Ranking")
plt.ylabel("Subscribers")
plt.title("Ranking vs Subscribers")
plt.show()
```

```python
#Most uploaded channel
top_uploads = df.sort_values("Uploads", ascending=False).head(10)

plt.bar(top_uploads["Username"], top_uploads["Uploads"])
plt.xticks(rotation=90)
plt.title("Channels with Most Uploads")
plt.show()
```

```python
#Subscribers vs Views
plt.figure(figsize=(10,6))

plt.scatter(df["Subscribers"], df["Views"], alpha=0.5)

plt.xscale("log")
plt.yscale("log")

plt.xlabel("Subscribers (log)")
plt.ylabel("Views (log)")
plt.title("Subscribers vs Views")

plt.show()
```

```python
plt.bar(df["Country"], df["Subscribers"])
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("Subscribers")
plt.title("Subscribers by Country")
plt.show()
```
